```python
import ast, json
import pandas as pd

pd.set_option("display.max_columns", None)

# load CSV summary
summary = pd.read_csv("outputs/recon_summary.csv")
summary["columns_different"] = summary["columns_different"].apply(
    lambda x: ast.literal_eval(x) if isinstance(x, str) and x.startswith("[") else x
)
display(summary.sort_values(["priority","cash_impact"], ascending=[False, False]).head(50))

# load JSONL details
records = [json.loads(line) for line in open("outputs/recon_details.jsonl","r",encoding="utf-8")]
detail_df = pd.json_normalize(records)
display(detail_df)

```
